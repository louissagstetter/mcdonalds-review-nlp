```python
# Datenanalyse & NLP bei McDonald's Reviews

# ============================================
# Phase 1 Daten aufbereiten und bereinigen
# ============================================

# ------------------------------------------
# 1. Bibliotheken importieren und NLTK vorbereiten
# ------------------------------------------
import pandas as pd
import nltk
# NLTK-Ressourcen herunterladen
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('omw-1.4')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import re


# Lemmatizer & Stopwords vorbereiten
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# ------------------------------------------
# 2. Datei einlesen
# ------------------------------------------
# Beispiel: CSV-Datei mit einer Textspalte "text"
# Passe den Dateipfad entsprechend an
file_path = "../data/mcd_reviews.csv"
df = pd.read_csv(file_path)

print("✅ Datei geladen. Vorschau:")
display(df.head())

# -------------------------------
# 3. Spalten checken
# -------------------------------
print("Spalten im DataFrame:")
print(df.columns)

# ------------------------------------------
# 4. Textbereinigung
# ------------------------------------------
def clean_text(text):
    text = str(text).lower()                     # alles klein schreiben
    text = re.sub(r"http\S+|www\S+|https\S+", "", text, flags=re.MULTILINE)  # URLs entfernen
    text = re.sub(r"[^a-zA-Z\s]", "", text)     # Sonderzeichen & Zahlen entfernen
    text = re.sub(r"\s+", " ", text).strip()    # Mehrfach-Leerzeichen entfernen
    return text

df['clean_text'] = df['review'].apply(clean_text)
print("✅ Texte bereinigt. Vorschau:")
display(df[['review', 'clean_text']].head())

# ------------------------------------------
# 5. Tokenisierung & Lemmatisierung
# ------------------------------------------
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [t for t in tokens if t.isalpha() and t not in stop_words]
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    return " ".join(tokens)

df['processed_text'] = df['clean_text'].apply(preprocess_text)
print("Texte verarbeitet (tokenisiert & lemmatisiert). Vorschau:")
display(df[['clean_text', 'processed_text']].head())

# -------------------------------
# 6. Häufigkeit der Wörter checken
# -------------------------------
from collections import Counter

all_tokens = " ".join(df['processed_text']).split()
word_freq = Counter(all_tokens)
print("Top 10 häufigste Wörter:")
print(word_freq.most_common(10))

# ------------------------------------------
# 7. Speichern der verarbeiteten Daten
# ------------------------------------------
output_file = "../data/verarbeitete_daten.csv"
df.to_csv(output_file, index=False, encoding='utf-8')
print(f"✅ Verarbeitete Daten gespeichert in {output_file}")



# ============================================
# Phase 2 Explorative Analyse (Wortfrequenzen)
# ============================================

# ------------------------------------------
# 1. Wortfrequenzen und Visualiserungen
# ------------------------------------------
from collections import Counter
import matplotlib.pyplot as plt

# Alle Tokens aus der Spalte 'processed_text' sammeln
all_tokens = " ".join(df['processed_text']).split()
word_freq = Counter(all_tokens)

# 20 häufigste Wörter extrahieren
common_words = word_freq.most_common(20)

# In DataFrame umwandeln
freq_df = pd.DataFrame(common_words, columns=['word', 'count'])

# Balkendiagramm zeichnen
plt.figure(figsize=(10,6))
plt.barh(freq_df['word'][::-1], freq_df['count'][::-1])
plt.title('Top 20 häufigste Wörter in McDonalds Reviews')
plt.xlabel('Anzahl')
plt.ylabel('Wort')
plt.show()


# ------------------------------------------
# 2. Stimmungsanalyse (Sentiment Analysis)
# ------------------------------------------

from textblob import TextBlob

# Sentiment berechnen (Polarity-Wert zwischen -1 = negativ und +1 = positiv)
df['sentiment'] = df['processed_text'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Verteilung der Stimmung anzeigen
plt.figure(figsize=(8,5))
plt.hist(df['sentiment'], bins=30, color='skyblue', edgecolor='black')
plt.title('Verteilung der Stimmung in Reviews')
plt.xlabel('Sentiment-Wert (-1 = negativ, +1 = positiv)')
plt.ylabel('Anzahl der Reviews')
plt.show()

# Durchschnitt pro Sternebewertung
df['rating_num'] = df['rating'].str.extract('(\d+)').astype(float)
sentiment_by_rating = df.groupby('rating_num')['sentiment'].mean()

print(sentiment_by_rating)

# ------------------------------------------
# 3. Themenextraktion (Topic Modeling mit LDA)
# ------------------------------------------

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Bag-of-Words erstellen
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
X = vectorizer.fit_transform(df['processed_text'])

# LDA-Modell trainieren (z. B. 5 Themen)
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(X)

# Top-Wörter je Thema anzeigen
terms = vectorizer.get_feature_names_out()
for idx, topic in enumerate(lda.components_):
    print(f"\nThema {idx+1}:")
    print(", ".join([terms[i] for i in topic.argsort()[-10:]]))



# ============================================
# Phase 3 Ergebnisse & Visualisierungen speichern
# ============================================

# ------------------------------------------
# 1. Themenmodellierung & Ergebnisse speichern
# ------------------------------------------

import os
import pandas as pd
import matplotlib.pyplot as plt
from gensim import corpora, models
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

# Sicherstellen, dass der Output-Ordner existiert
output_dir = "outputs"
os.makedirs(output_dir, exist_ok=True)

# ------------------------------------------
# 2. LDA-Modell trainieren
# ------------------------------------------

texts = [t.split() for t in df['processed_text']]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

lda_model = models.LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=5,
    random_state=42,
    passes=10,
    alpha='auto'
)

print("✅ LDA-Modell trainiert.")

# ------------------------------------------
# 3. LDA Visualisierung speichern
# ------------------------------------------

lda_vis = gensimvis.prepare(lda_model, corpus, dictionary)
pyLDAvis.save_html(lda_vis, os.path.join(output_dir, "lda_visualization.html"))
print(f"✅ LDA-Visualisierung gespeichert in {output_dir}/lda_visualization.html")

# ------------------------------------------
# 4. Themen als CSV speichern
# ------------------------------------------

topic_words = []
for i, topic in enumerate(lda_model.show_topics(num_topics=5, num_words=10, formatted=False)):
    topic_words.append({
        "Thema": f"Thema {i+1}",
        "Top_Wörter": ", ".join([word for word, _ in topic[1]])
    })

topics_df = pd.DataFrame(topic_words)
topics_df.to_csv(os.path.join(output_dir, "lda_topics.csv"), index=False)
print(f"✅ Themen als CSV gespeichert in {output_dir}/lda_topics.csv")

# ------------------------------------------
# 5. Top 20 Wörter (Plot) speichern
# ------------------------------------------

top_words = pd.Series([word for text in texts for word in text]).value_counts().head(20)
plt.figure(figsize=(10,6))
top_words.plot(kind='bar')
plt.title("Top 20 häufigste Wörter")
plt.ylabel("Häufigkeit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "top20_words.png"))
plt.show()
print(f"✅ Top 20 Wörter Plot gespeichert in {output_dir}/top20_words.png")

# ------------------------------------------
# 6. Sentiment vs Rating Plot speichern
# ------------------------------------------

plt.figure(figsize=(8,5))
df.groupby('rating_num')['sentiment'].mean().plot(kind='bar', color='skyblue')
plt.title("Durchschnittliche Sentiments nach Bewertung")
plt.ylabel("Durchschnittlicher Sentimentwert")
plt.xlabel("Bewertung")
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "sentiment_vs_rating.png"))
plt.show()
print(f"✅ Sentiment vs Rating Plot gespeichert in {output_dir}/sentiment_vs_rating.png")

# ------------------------------------------
# 7. Interaktive LDA-Visualisierung (pyLDAvis)
# ------------------------------------------

import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
import matplotlib.pyplot as plt

# Visualisierung erstellen
lda_vis = gensimvis.prepare(lda_model, corpus, dictionary)

# Ordner sicherstellen
os.makedirs("../outputs", exist_ok=True)

# HTML-Dashboard speichern
pyLDAvis.save_html(lda_vis, "../outputs/lda_visualization.html")

# Hinweis: erfordert zusätzliches Paket 'selenium' + 'pyppeteer' oder 'imgkit'
# Für einfache statische Alternative rendern wir ein einfaches Plotbild:

fig, ax = plt.subplots(figsize=(10, 6))
for i, topic in enumerate(lda_model.show_topics(num_topics=5, num_words=10, formatted=False)):
    words = [w for w, _ in topic[1]]
    weights = [v for _, v in topic[1]]
    ax.barh(words, weights, label=f'Topic {i+1}', alpha=0.6)
ax.legend()
ax.set_title("LDA Themenübersicht (vereinfachte Darstellung)")
plt.tight_layout()
plt.savefig("../outputs/lda_visualization.png")
plt.close()

print("✅ pyLDAvis-Dashboard und Themenabbildung wurden in '../outputs/' gespeichert.")
```
